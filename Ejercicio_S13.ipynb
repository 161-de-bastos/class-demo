## Azure VM

1. Cree una máqina virtual en Azure

- La creación de los recursos, realizarlo en un grupo de recursos.
- Características de máquina de bajo costo:
  - Security type: Standard
  - VM architecture: Arm64
  - Image: Ubuntu Server 24.04 LTS – ARM64 Gen2
  - Size: Standard_B2pls_v2 – 2 vcpus, 4GiB memory (24,53 US$/month)
- Verificar que se pueda conectar desde el exterior por el puerto SSH (22)
- Descargar y guardar las credenciales de conexión (file.pem)

## Comandos Linux

2. Cree un programa y ejecute en la VM de Azure:

- Crear directorios llamados: dir_1, dir_2, ..., dir_5
- Crear directorio llamado: destino
- Crear 1000 archivos en dir_1, ejemplo: archivo_1.txt, archivo_2.txt, ... archivo_1000.txt
- Mover los archivos al directorio 'destino'
- Mostar el total de archivos movidos

### Aqui coloque el archivo bash
